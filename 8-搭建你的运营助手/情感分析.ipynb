````python
import openai

def get_completion(prompt):
    # 设置OpenAI API的访问密钥
    #openai.api_key = 'XXXX'
    
    # 使用OpenAI GPT-3.5模型生成完成文本
    response = openai.Completion.create(
        engine='text-davinci-003',  # 使用GPT-3.5模型
        prompt=prompt,
        max_tokens=100,  # 生成的最大标记数
        temperature=0.7,  # 控制生成文本的多样性，值越大则越随机
        n = 1,  # 生成1个输出
        stop=None,  # 可以指定停止生成的标记序列
        echo=True  # 返回包含输入和输出的完整响应对象
    )
    
    # 提取生成的文本
    completion = response.choices[0].text.strip()
    return completion

review = '这款音效特别好 给你意想不到的音质。'
prompt = f"""
下面产品评论的情况是怎样的，产品评论使用 ``` 进行表示
产品评论：```{review}```
"""
response = get_completion(prompt)
print(response)

prompt = f"""
下面产品评论的情况是怎样的，产品评论使用 ``` 进行表示
回复请用一个词语：正向 或者 负向
产品评论：```{review}```
"""
response = get_completion(prompt)
print(response)
````
